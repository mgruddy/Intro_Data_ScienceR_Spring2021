Recall
- **Factors** are *augmented* vectors, i.e. vectors with additional meta-data
- The **levels** are the unique values a factor vector is allowed to take on, which are also ordered

````r
# the possible values a vector of months can take on, ordered chronologically
months = c("Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sept", "Oct", "Nov", "Dec")

# create a factor in this way
summer_months <- factor(c("Jun", "Jul", "Aug", "jam"), levels = months)
summer_months
````

````r
# which variables can/should be factors?
mpg
````

````r
mpg %>%
    mutate(cyl = factor(cyl))
````

````r
factor(mpg$class)
````

````r
# how should we order this factor?
unique(mpg$class)
````

````r
mpg %>%
    mutate(class = factor(class, levels = c("2seater", "subcompact", "compact", 
                                            "midsize", "suv", "minivan", "pickup")))
````

````r
# the factor is mapped to an aesthetic via its order
mpg %>%
    mutate(class = factor(class, levels = c("2seater", "subcompact", "compact", 
                                            "midsize", "suv", "minivan", "pickup"))) %>%
    ggplot(aes(x = hwy, y = class)) +
    geom_point()
````

````r
# we can reorder a factor by another variable as well in ggplot
class_mean <- mpg %>%
    mutate(class = factor(class)) %>%
    group_by(class) %>%
    summarize(mean_hwy = mean(hwy, na.rm = TRUE)) %>%
    arrange(mean_hwy, desc=TRUE) %>%
    print()
````

````r
# reorder by mean_hwy
class_mean %>%
    ggplot(aes(x = mean_hwy, y = fct_reorder(class, mean_hwy, .desc = TRUE))) +
    geom_point()
````

````r
# for bar graphs we can use fct_infreq
mpg %>%
    mutate(class = factor(class)) %>%
    ggplot() +
    geom_bar(aes(x = fct_infreq(class)))
````

````r
# we can rename factors easily with fct_recode
mpg %>%
  mutate(drv = fct_recode(drv,
    "front"       = "f",
    "rear"        = "r",
    "four-wheel"  = "4"))
````

````r
mpg %>%
    mutate(class = fct_collapse(class,
            car =   c("2seater", "subcompact", "compact", "midsize"),
            van =   c("suv", "minivan"),
            truck = c("pickup")))
````

````r
# we can also collapse levels of factors for a "broader" view
mpg %>%
    count(class)

mpg %>%
    mutate(class = fct_collapse(class,
            car =   c("2seater", "subcompact", "compact", "midsize"),
            van =   c("suv", "minivan"),
            truck = c("pickup"))) %>%
    count(class)
````

````r
# fct_lump collapses all the smaller levels together
# the 3 largest levels are kept, and the rest are collapsed to "Other"
mpg %>%
    mutate(class = fct_lump(class, n = 3)) %>%
    count(class)
````
