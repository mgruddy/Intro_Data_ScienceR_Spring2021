````r
ggplot(data = mpg) + 
  geom_bar(mapping = aes(x = class, fill = drv))
````

The ```position``` option for ```geom_bar``` alters the bar graph in interesting ways.
 - The default is ```stack``` which stacks the bars on top of each other.
 - The option ```identity``` overlaps the bars.
 - The option ```dodge``` places the bars next to one another.
 - The option ```fill``` makes each bar the same height.

````r
ggplot(data = mpg) + 
  geom_bar(mapping = aes(x = class, alpha = drv), position = "dodge")
````

Run the following block of code with and without the option ```alpha = .5```

````r
ggplot(data = mpg) + 
  geom_bar(mapping = aes(x = class, fill = drv), alpha = .5, position = "identity")
````
