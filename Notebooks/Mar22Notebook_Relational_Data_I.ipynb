Different datasets relate to each other via variables or groups of variables called **keys**
- A **primary key** uniquely identifies an observation in its *own* table
- A **foreign key** uniquely identifies an observation in *another* table
- A primary key and the corresponding foreign key together form a **relation**

````r
# In the airlines dataset, "carrier", "name", and ("carrier", "name") are primary keys
# "carrier" is a foreign key for the flights dataset
airlines
````

A **surrogate key** is a variable (or group of variables) created for the purpose of being a primary or foreign key.

````r
# Here we create a surrogate key using the ordering of the rows
flights %>%
    mutate(index = row_number()) %>%
    select(index, everything())    # moves index column to first column
````

### Types of joins

- A **mutating join** adds a new variable (hence "mutating") to a dataset from matching observations in another dataset. 
 * Inner join: Observations are matched whenever the keys are equal
 * left/right joins: For one of the datasets, observations are matched to each key in that dataset and missing values inserted where necessary
 * full join: Each observation in both datasets gets matched along the keys, missing values inserted where necessary
- A **filtering join** filters observations from one data frame based on whether or not they match an observation in the other table.
 * semi-join: keeps all observations in a table who have a match in another table
 * anti-join: drops all observations in a table who have a match in another table
- **Set operations** combine datasets with the same variable sets, treating them as sets of observations
