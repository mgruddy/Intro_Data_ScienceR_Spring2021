Other column functions:
- ```mean(x)```: computes a number, the mean of column x
- ```sum(x)```: computes a number, the sum of a column x
- ```lead(x)```, ```lag(x)```: computes a new column, shifted down or up repsectively
- cumulative functions such as ```cumsum``` and ```cummean```: computes a new column, with the cumulative sum, mean, etc.
- ranking functions such as ```min_rank```: computes a new column, with the ranking of a column (2nd, 1st, etc.)

````r
# adjust by the mean
print(transmute(flights,
                delay = dep_delay - mean(dep_delay)))
````

````r
# remove missing values
print(transmute(flights,
                delay = dep_delay - mean(dep_delay, na.rm = TRUE)))
````

````r
# rank the flights by the least delayed
print(mutate(flights,
                delay_rank = min_rank(dep_delay)), width = 500)
````

````r
# rank the flights by the most delayed
print(mutate(flights,
                delay_rank = min_rank(desc(dep_delay))), width = 500)
````

To perform more than one operation on a dataset we can use the pipe, ```%>%```, to simplify our code.

Task: Create a dataset of the ten fastest flights.

````r
flight_speeds <- mutate(flights, 
                            air_time_hours = air_time / 60, 
                            avg_speed_mph = distance / air_time_hours)

flight_speeds_ranked <- mutate(flight_speeds, avg_speed_rank = min_rank(desc(avg_speed_mph)))

(ten_fastest <- filter(flight_speeds_ranked, avg_speed_rank <= 10))
````

````r
ten_fastest <- flights %>%
  mutate(air_time_hours = air_time / 60, avg_speed_mph = distance / air_time_hours) %>%
  mutate(avg_speed_rank = min_rank(desc(avg_speed_mph))) %>%
  filter(avg_speed_rank <= 10)
````
