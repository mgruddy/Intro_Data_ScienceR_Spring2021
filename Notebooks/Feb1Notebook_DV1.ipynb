````r
# plot engine volume against highway mpg
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = cty))
````

1. Plot two different columns against one another
2. Run just the line ```ggplot(data = mpg)```
3. Think about how we might be able to visualize more than two columns on a 2D plot

- The code ```ggplot(data = mpg)``` creates the base plot and indicates that we will be drawing from the ```mpg``` dataframe for our data.
- To the base plot we add a ```geom_point``` layer, which creates a coordinate system and scatter plot based on the mapping we give it.
- An *aesthetic* is a visual property of objects in your plot like the color, shape, or size of each data point.

````r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy, alpha = cty, color = class))
````

1. Try out the aesthetic options color, size, shape, and alpha. What do each do?
2. How could we find out more about the available aesthetics for the ```geom_point``` layer?

Be careful with writing code across multiple lines!

````r
# okay
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy, color = cty))

# not okay
ggplot(data = mpg)
  + geom_point(mapping = aes(x = displ, y = hwy, color = cty))

# okay
ggplot(data = mpg) + 
  geom_point(
    mapping = aes(x = displ, y = hwy, color = cty))

# not okay
ggplot(data = mpg) + 
  geom_point
    (mapping = aes(x = displ, y = hwy, color = cty))
````

Run the following two blocks of code. What happened and why are they different?

````r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy), color = "blue")
````

````r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy, color = "blue"))
````

Another way to visualize more than 2 variables is create side-by-side scatter plots

````r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy)) + 
  facet_wrap(~ class, nrow = 2) # nrow is an optional argument!
````

````r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy)) + 
  facet_grid(drv ~ class)
````

1. Try out a few different facet grids!
